```python
facts = {"p"}

rules = [
    ("p", "q"),
    ("q", "r"),
    ("r", "s")
]

goal = "D"

while True:
    new_fact_added = False

    for condition, result in rules:
        if condition in facts and result not in facts:
            print(f"Applying rule: {condition} -> {result}")
            facts.add(result)
            new_fact_added = True

            if result == goal:
                print("Goal reached:", goal)
                break

    if goal in facts or not new_fact_added:
        break
    print("Final Facts:", facts)
```

```python

rules = [
    ("A", "B"),
    ("B", "C"),
    ("C", "D")
]

facts = {"A"}

def backward_chain(goal):
    if goal in facts:
        return True

    for condition, result in rules:
        if result == goal:
            print(f"Checking rule: {condition} -> {result}")
            if backward_chain(condition):
                return True

    return False


goal = "D"

if backward_chain(goal):
    print("Goal reached:", goal)
else:
    print("Goal not reached:", goal)
```

```python
facts = {"A"}

rules = [
    ("A", "B"),
    ("B", "C"),
    ("C", "D")
]

goal = "D"

step = 1

while True:
    print(f"\n--- Iteration {step} ---")
    print("Current facts:", facts)

    new_fact_added = False

    for condition, result in rules:
        print(f"Checking rule: {condition} -> {result}")

        if condition in facts and result not in facts:
            print(f" Applying rule: {condition} -> {result}")
            facts.add(result)
            print(" Updated facts:", facts)

            new_fact_added = True

            if result == goal:
                print("\n Goal reached:", goal)
                break

    if goal in facts:
        break

    if not new_fact_added:
        print("\n No new facts can be added. Stopping.")
        break

    step += 1

print("\nFinal Facts:", facts)
```
